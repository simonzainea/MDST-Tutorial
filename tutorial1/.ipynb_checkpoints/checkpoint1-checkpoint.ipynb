<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

(Feel free to import other libraries!)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df = pd.read_csv('../data/states_edu.csv')
df.shape
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
df.isna().sum()
df.notna().sum()
df.dropna()
df.dropna(subset=['AVG_MATH_4_SCORE'], inplace=True)
df.head()
```

<h2>Exploratory Data Analysis (EDA) </h2>

Chosen one of Grade 4 Reading, Grade 4 Math, or Grade 8 Math to focus on: *Grade 4 Math*

How many years of data are logged in our dataset? 

```python
max_years = df['YEAR'].max()
min_years = df['YEAR'].min()
years = max_years - min_years
print(years)
```

Let's compare Michigan to Ohio. Which state has the higher average across all years in the test you chose?

```python
mich_data = df[df['STATE'] == 'MICHIGAN']
ohio_data = df[df['STATE'] == 'OHIO']
mich_mean = mich_data['AVG_MATH_8_SCORE'].mean()
ohio_mean = ohio_data['AVG_MATH_8_SCORE'].mean()
if mich_mean > ohio_mean:
    print('Michigan has the higher average 8th grade scores:', mich_mean)
elif ohio_mean > mich_mean:
    print('Ohio has the higher average 8th grade scores:', ohio_mean)
else:
    print('Michigan and Ohio have the same average 8th grade test scores')
```

Find the average for your chosen test across all states in 2019

```python
year_2019 = df[df['YEAR'] == 2019]
avg_math_4_test = year_2019['AVG_MATH_8_SCORE'].mean()
print(avg_math_4_test)
```

For each state, find a maximum value for your chosen test score

Refer to the `Grouping and Aggregating` section in Tutorial 0 if you are stuck.

```python
states = df.groupby('STATE')
max_math = states['AVG_MATH_8_SCORE'].max()
print(max_math)
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df['INSTRUCTION_EXPENDITURE_PER_STUDENT'] = df['INSTRUCTION_EXPENDITURE'] / df['GRADES_ALL_G']
df['INSTRUCTION_EXPENDITURE_PER_STUDENT'].dropna()
```

Feature engineering justification: I created a new column that considers instructional funding per each student for each state. I made this feature because I have a prediction that a the more money put towards instruction will result in higher 8th grade math scores.

<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.groupby('AVG_MATH_8_SCORE')["INSTRUCTION_EXPENDITURE_PER_STUDENT"].mean().plot()
plt.ylabel('Instructional Exependiture per Student')
plt.xlabel('Average 8th Grade Math Score')
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['AVG_MATH_4_SCORE','INSTRUCTION_EXPENDITURE_PER_STUDENT']].dropna()
y = y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_MATH_4_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'INSTRUCTION_EXPENDITURE_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_4_SCORE')
plt.title("Model Behavior on Testing Set")
```
